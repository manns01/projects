```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

t_ = np.linspace(0, 20, 200)
omega1, omega2, omega3 = np.pi/4, np.pi, 2*np.pi
print(omega1/(2*np.pi), omega2/(2*np.pi), omega3/(2*np.pi)) ## using omega = 2*pi*f
func = np.sin(omega1*t_)+np.sin(omega2*t_)+np.sin(omega3*t_)
plt.plot(t_, func, '-')
plt.show()

fft_func = np.fft.fft(func)
#plt.plot(t_, fft_func.real)
#plt.show()

N_func = func.size
dt_func = (max(t_) - min(t_))/t_.size
freq_func = np.fft.fftfreq(N_func, dt_func)
#plt.plot(t_, freq_func.real)
#plt.show()

## power spectrum
ps_func = np.abs(fft_func)**2
plt.plot(freq_func, ps_func, '--', label='Power spectrum')
plt.legend()
plt.show()
peaks, _ = find_peaks(ps_func)
peak_freqs = freq_func[peaks]
print(peak_freqs)
print(rf'omega1 = {2*peak_freqs[0]} pi, omega2 = {2*peak_freqs[1]} pi, omega3 = {2*peak_freqs[2]} pi')
```
