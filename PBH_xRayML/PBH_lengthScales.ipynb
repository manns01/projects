```python
import numpy as np
import matplotlib.pyplot as plt

c = 3.e8 ## m/s
G = 6.6743e-11 
M_sun = 1.9891e30 ## kg
M1 = 1.e-16*M_sun
M2 = 1.e-12*M_sun
R_schw1 = (2*G*M1/c**2)*5.29e11
R_schw2 = (2*G*M2/c**2)*5.29e11
R_atom = 1.0 ## Hydrogen atom size (Bohr radius) = 52.9 pm = 5.29e-11 m
print(R_schw1, R_schw2)
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, ax = plt.subplots(figsize=(8, 8))

R_min = R_atom
R_max = R_schw2
R_min_new = R_schw1
R_max_new = R_atom

x_center = 0
    
circle1 = plt.Circle((x_center, 0), R_min, color='black', alpha=1.0, fill=True)  # Filled with blue color
ax.add_patch(circle1)

circle2 = plt.Circle((x_center, 0), R_max, color='mediumpurple', alpha = 0.2, fill=True)  # Filled with blue color
ax.add_patch(circle2)

ax.set_aspect('equal', adjustable='box')

ax.set_xlim(-R_max - 1, R_max + 1)
ax.set_ylim(-R_max - 1, R_max + 1)

## inset for corner plot
axin = inset_axes(ax, width="30%", height="30%", loc='lower right')
axin.set_aspect('equal')
axin.set_xlim(-R_max_new, R_max_new)
axin.set_ylim(-R_max_new, R_max_new)

circle3 = plt.Circle((x_center, 0), R_min_new, color='navy', alpha=0.5, fill=True)
axin.add_patch(circle3)

circle4 = plt.Circle((x_center, 0), R_max_new, color='cornflowerblue', alpha=0.2, fill=True)
axin.add_patch(circle4)

ax.set_title(r'PBH length scale visualization')
ax.set_xlabel(r'x (in units of $a_o = 52.9$ pm)')
ax.set_ylabel(r'y (in units of $a_o = 52.9$ pm)')
axin.set_title(r'zoomed view of H-atom')

## annotations
ax.text(-400, 600, r'Primordial black hole')
ax.text(-340, 510, r'($M_{PBH} = 10^{-12} M_\odot$)')
ax.text(-650, -25, r'Hydrogen atom$\rightarrow$')

axin.text(-0.5, 0.4, r'Hydrogen atom')
axin.text(-0.1, -0.15, r'$\uparrow$', ha='left', va='top')
axin.text(-0.7, -0.4, r'Primordial black hole')
axin.text(-0.6, -0.6, r'($M_{PBH} = 10^{-16} M_\odot$)')

##plt.savefig('PBH_atom_fig.pdf', dpi=300)
plt.show()
```
